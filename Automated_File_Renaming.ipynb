# A Notebook to automate the file renaming process 

```python
# importing the modules
import os
import shutil
import re
import osgeo
import rasterio
# from rasterio import shutil
```

```python
# Adding the count to the new file name and extension
folder = r'D:\oilspill_dataset\train\images\\'
count = 1 
images= []
try:
    for x in list(range(1,56)):
        origin = r'D:\Disseration_Data\OilSpills\oilspill_'+str(x).zfill(2)
        for file in os.listdir(origin):
                if re.fullmatch('.+px.tif', file):
                    print('{} to {}'.format(origin + '\\'+ file, folder))
#                     print('{} to {}'.format(origin + '\\'+ file, folder + "image_" + str(count).zfill(2) + ".tif"))
                    shutil.copy2(origin + '\\'+ file, folder)
                    images.append(file)
                    count += 1
except FileNotFoundError as e:
    print(e)

print("Files are copied successfully")
print(len(images))
```

```python
# TO COPY FILES
folder = r'D:\oilspill_dataset\train\wind\\'
count = 1

masks = []
for x in list(range(1,56)):
    origin = r'D:\Disseration_Data\OilSpills\oilspill_'+str(x).zfill(2)
    for file in os.listdir(origin):
            if re.fullmatch('.+wind_speed_verde.+tif', file):
                print('{} to {}'.format(origin + '\\'+ file, folder))
                shutil.copy2(origin + '\\'+ file, folder)
                
#                 print('{} to {}'.format(origin + '\\'+ file, folder + "mask_" + str(count).zfill(2) + ".tif"))
#                 rasterio.shutil.copy(origin + '\\'+ file, folder + "mask_" + str(count).zfill(2) + ".tif")
                masks.append(file)
                count += 1
print(len(masks), "Files are copied successfully")
```

```python
# TO RENAME TRAINING IMAGES
folder = r'D:\oilspill_dataset\train\masks\\'
count = 1
# count increase by 1 in each iteration
# iterate all files from a directory

for file_name in os.listdir(folder):
    print('old file name: {}'.format(file_name))
    # Construct old file name
    source = folder + file_name

    # Adding the count to the new file name and extension
    destination = folder + str(count) + "_" + file_name

    # Renaming the file
    os.rename(source, destination)
    print('new file name: {}'.format(destination))
    print('------')
    count += 1
print('All Files Renamed')

print('New Names are')
# verify the result
res = os.listdir(folder)
print(res)
```
